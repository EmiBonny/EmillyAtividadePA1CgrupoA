```python
# Peça uma palavra e conte as letras: ex: python - a palavra tem 06 letras
word = str (input("Digite uma palavra: "))
quantidade = len(word)
print(f"A palavra {word} tem {quantidade:02d} letras")
for i in range (0, quantidade + 1):
  print (i)
```

```python
# Peça uma palavra e conte as vogais
word = str (input("Digite uma palavra: "))
vogais = "aeiou"
contador = 0

for letra in word:
    if letra in vogais:
        contador += 1

print(f"A palavra {word} tem {contador} vogais")
```
